```python

CC_LIST = ["봉의동","요선동","낙원동""중앙로1가""중앙로2가""중앙로3가""옥천동""조양동""죽림동""운교동""약사동""효자동""소양로1가""소양로2가"
           "소양로3가""소양로4가""근화동""우두동""사농동""후평동""온의동""교동""퇴계동""석사동""삼천동""칠전동""송암동""신동""중도동"
           "신북읍""동면""동산면""신동면""남면""서면""사북면""북산면""동내면"
           "남산면""조운동""약사명동""소양동""후평1동""후평2동""후평3동""효자1동""효자2동""효자3동""강남동""신사우동"]
food=['한식집', '분식집', '카페', '파스타']
service=['네일','미용실']
retail=['약국', '편의점', '화장품', '슈퍼마켓','인테리어','애완동물','꽃집','가구','조명']

# 웹 드라이버 초기화
driver = webdriver.Chrome(executable_path='./chromedriver')

# 크롤링 대상 키워드 목록
crawling_list = []
for location in CC_LIST:
    for category in retail:
        crawling_list.append(location + " " + category)       

# 리뷰 수집 함수
def collect_reviews(keyword):
    # 네이버 지도 페이지로 이동
    # driver.get(f'https://map.naver.com/v5/search/{keyword}?c=15.69,0,0,0,dh')
    url = f'https://map.naver.com/v5/search/{keyword[0]}%20{keyword[1]}?c=13.00,0,0,0,dh'
    driver.get(url)
            
    # 스크롤 내리기
    for _ in range(2):  # 예시로 2번 스크롤을 내리도록 설정
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # 스크롤 후 로딩 대기 시간
    
        # 리뷰 칸에 들어가기
        cate_pls = driver.find_elements(By.CLASS_NAME, 'place_list_item')
        for j in range(len(cate_pls) - 1):  # 전체 길이로 지정하면 안됨
            if cate_pls[j].text == "리뷰":
                cate_pls[j].click()
            else:
                if j == len(cate_pls):
                    break
                else:
                    pass

        # 더보기 클릭 (최대 10번)
    for _ in range(10):
        try:
            driver.find_element(By.CLASS_NAME, 'fvwqf').click()
        except:
            break
            
    # 리뷰 수집
    reviews = []
    review_elements = driver.find_elements(By.CLASS_NAME, 'xHaT3')
    for review_element in review_elements:
        review_text = review_element.text
        reviews.extend(review_text.split())
    
    return reviews

# 리뷰 수집 및 저장
for keyword in crawling_list:
    try:
        reviews = collect_reviews(keyword)
        print(f'Keyword: {keyword}, 리뷰 수: {len(reviews)}')
        # 여기서 리뷰 데이터를 저장 또는 처리할 수 있습니다.
    except Exception as e:
            print(f'Keyword: {keyword}에서 리뷰 수집 중 오류 발생: {str(e)}')

# 웹 드라이버 종료
driver.quit()
```
